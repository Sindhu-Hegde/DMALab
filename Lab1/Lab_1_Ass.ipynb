```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

train_data = pd.read_csv("train_users_2.csv")
test_data = pd.read_csv("test_users.csv")

display(train_data.count())
display(train_data.head().transpose())
train_data.shape[0]
```

```python
df_all = pd.concat((train_data, test_data), ignore_index = True)
df_all.head().transpose()
```

```python
df_all.count()
```

```python
df_all["timestamp_first_active"] = pd.to_datetime(df_all["timestamp_first_active"], format = "%Y%m%d%H%M%S")
df_all.head().transpose()
```

```python
display(sum(df_all["date_account_created"].isnull().values.ravel()))
display(sum(df_all["age"].isnull().values.ravel()))
display(sum(df_all["date_first_booking"].isnull().values.ravel()))
```

```python
df_all.drop("date_first_booking", axis = 1, inplace = True)
```

```python
df_all.head().transpose()
```

## To fill in the missing values of age:

1] If the gender for the missing age known, then the age can be filled by taking the mean of ages of either male or female depending on the gender.

2] If the gender is also unknown, then fill age by some constant value (like 0 or -1)

```python
import numpy as np

df_all.gender.replace('-unknown-', np.nan, inplace=True)
df_all.head().transpose()
```

```python
import seaborn as sns
%pylab inline

df_all.signup_app.value_counts(dropna=False).plot(kind='bar', color='red', rot=0)
plt.xlabel('Signup_App')
sns.despine()
```

## Summary:

Most of the users prefer "Web" as the signup_app


```python
df_all.signup_method.value_counts(dropna=False).plot(kind='bar', color='blue', rot=0)
plt.xlabel('Signup_Method')
sns.despine()
```

## Summary:

Most of the users prefer the "basic" sign-up method

```python
df_all.first_device_type.value_counts(dropna=False).plot(kind='bar', color='green', rot=0)
plt.xlabel('Device_Type')
sns.despine()
```

```python
df_all.language.value_counts(dropna=False).plot(kind='bar', color='purple', rot=0)
plt.xlabel('Language')
sns.despine()
```

## Summary :

Most of the users prefer 'English' language.
